```python
applyurl = soup.findAll('a', text=('Apply'))
```

```python
applyurl0 = [x.get('href') for x in applyurl]
```

```python
print(applyurl0)
```

```python
dataframe01.insert(4, 'Apply URL', applyurl0)
```

```python
desc_list = []
```

```python
for x in applyurl0:
    xyz = requests.get(x)
    desc001 = bs(xyz.text)
    description01 = desc001.findAll('p', class_=None, id=None)
    new_desc = description01[0].text
    desc_list.append(new_desc)
```

```python
desc_list
```

```python
dataframe01.insert(5, 'Description', desc_list)
```
